```python
#INTERPOLATION
```

```python
from scipy.interpolate import interp1d
import numpy as np
xs = np.arange(10)
ys = 2*xs + 1
interpolation_func = interp1d(xs,ys)
d = interpolation_func(np.arange(2.1 ,3 ,0.1))
print(d)
```

```python
#SPLINE INTERPOLATION
```

```python
from scipy.interpolate import UnivariateSpline
import numpy as np
xs = np.arange(10)
ys = xs** + np.sin(xs) + 1

interpolation_func = UnivariateSpline(xs,ys)
d = interpolation_func(np.arange(2.1 ,3 ,0.1))
print(d)
```

```python
#INTERPOLATION WITH RADIAL BASIS FUNCTION
```

```python
from scipy.interpolate import Rbf
import numpy as np
xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1
interpolation_func = Rbf (xs, ys)
d = interpolation_func(np.arange(2.1 ,3 ,0.1))
print(d)
```
