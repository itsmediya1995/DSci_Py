```python
#Delaunay 
```

```python
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt
d = np.array([[2,4],[3,4],[3,0],[2,2],[4,1]])
simplices = Delaunay(d).simplices
plt.triplot(d[:,0],d[:,1],simplices)
plt.scatter(d[:,0],d[:,1],color = 'r')
plt.show()
```

```python
#Convex Hull
```

```python
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt
d = np.array([[2,4],[3,4],[3,0],[2,2],[4,1],[1,2],[5,0],[3,1],[1,2],[0,2]])
hull = ConvexHull(d)
hull_points = hull.simplices
plt.scatter(d[:,0],d[:,1])
for simplex in hull_points:
    plt.plot(d[simplex,0],d[simplex,1],'k-')
plt.show()
```

```python
#KDTrees
```

```python
from scipy.spatial import KDTree
d= [(1,-1),(2,3),(-2,3),(2,-3)]
tree = KDTree(d)
res = tree.query((1,1))
print(res)
```

```python
#Distance matrix , 2types : [Euclidean & Cosine] distance
```

```python
#Euclidean
from scipy.spatial.distance import euclidean
p1 = (1,0)
p2 = (10,2)
d = euclidean(p1,p2)
print(d)
```

```python
#cosine
from scipy.spatial.distance import cosine
p1 = (1,0)
p2 = (10,2)
d = cosine(p1,p2)
print(d)
```
