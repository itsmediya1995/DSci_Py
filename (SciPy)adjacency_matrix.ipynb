```python
import numpy as np
from scipy.sparse.csgraph import connected_components
from scipy.sparse import csr_matrix
d = np.array([[0,1,2],[1,0,0],[2,0,0]])
dNew = csr_matrix(d)
print (connected_components(dNew))
```

```python
#Dijkstra method
```

```python
import numpy as np
from scipy.sparse.csgraph import dijkstra
from scipy.sparse import csr_matrix
d = np.array([[0,1,2],[1,0,0],[2,0,0]])
dNew = csr_matrix(d)
print (dijkstra(dNew , return_predecessors = True , indices = 0))
```

```python
#Floyd Warshall
```

```python
import numpy as np
from scipy.sparse.csgraph import floyd_warshall
from scipy.sparse import csr_matrix
d = np.array([[0,1,2],[1,0,0],[2,0,0]])
dNew = csr_matrix(d)
print (floyd_warshall(dNew , return_predecessors = True))
```

```python
#Bellman_Ford
```

```python
import numpy as np
from scipy.sparse.csgraph import bellman_ford
from scipy.sparse import csr_matrix
d = np.array([[0,-1,2],[1,0,0],[2,0,0]])
dNew = csr_matrix(d)
print (bellman_ford(dNew , return_predecessors = True , indices = 0))
```

```python
#Depth_First_Order
```

```python
import numpy as np
from scipy.sparse.csgraph import depth_first_order
from scipy.sparse import csr_matrix
d = np.array([[0,1,0,1],[1,1,1,1],[2,1,1,0],[0,1,0,1]])
dNew = csr_matrix(d)
print (depth_first_order(dNew , 1))
```

```python
#breadth_first_order
```

```python
import numpy as np
from scipy.sparse.csgraph import breadth_first_order
from scipy.sparse import csr_matrix
d = np.array([[0,1,0,1],[1,1,1,1],[2,1,1,0],[0,1,0,1]])
dNew = csr_matrix(d)
print (breadth_first_order(dNew , 1))
```
