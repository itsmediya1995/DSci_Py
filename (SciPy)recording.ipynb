```python
import sounddevice
from scipy.io.wavfile import write
def dvoice_recorder(seconds ,file):
    print("recording started ... ")
    drecording = sounddevice.rec((seconds*44100),samplerate = 44100 , channels = 2)
    sounddevice.wait()
    write(file , 44100 , drecording)
    print ("Recording is finished")
    dvoice_recorder(10,"recorder.wav")
```
